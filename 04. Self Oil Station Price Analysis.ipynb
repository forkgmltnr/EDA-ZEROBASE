# 16_실습 - 주요소 가격 정보 정리하기

````python
import pandas as pd
from glob import glob
````

````python
#glob으로 파일 목록 한번에 가져온다.
````

````python
glob("../data/지역_*.xls")
````

````python
# 파일명 저장
stations_files = glob("../data/지역_*.xls")
stations_files[:5]
````

````python
#하나만 읽어오기
#hedaer 매개변수를 사용하여 엑셀 파일의 컬럼을 설정 한다.
#header =2 사용하여 컬럼명이 제대로 출력된다. (excel 파일에서 컬럼들을 확인해주기)
tmp = pd.read_excel(stations_files[0], header=2)
tmp.tail(2)
````

````python
tmp_raw = []
for file_name  in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
````

````python
#형식이 동일하고 연달아 붙이기만 하면 될 때는 concat 사용
#concat을 사용하여 위에서 읽어드린 주유소 정보를 데이터 프레임으로 만든다.
#491개 주유소 확인
stations_raw = pd.concat(tmp_raw)
stations_raw
````

````python
stations_raw.info()
````

````python
stations_raw.columns
````

````python
stations = pd.DataFrame({
    "상호": stations_raw["상호"],
    "주소": stations_raw["주소"],
    "가격": stations_raw["휘발유"],
    "셀프": stations_raw["셀프여부"],
    "상표": stations_raw["상표"]
    
})
stations.tail()
````

````python
for eachAddress in stations["주소"]:
    print(eachAddress.split()[1])
````

````python
# for문과 split를 사용하여 주소에 있는 자치구를  station["구"] 컬럼에 추가하고 넣어주자
#stations 주소 컬럼에서 split 공백 기준으로 분할한다.
#분할된 결과 중 인덱스1 요소를 출력하여 "구"열에 추가
stations["구"] = [eachAddress.split()[1] for eachAddress in stations["주소"]]
stations
````

````python
stations["구"].unique(), len(stations["구"].unique())
````

````python
# 가격 정보 없는 주유소 출력 및  미사용
#가격 정보 없는 주유소 확인해주기
stations[stations["가격"]=="-"]
````

````python
#가격 정보가 있는 주유소만 사용하기
stations = stations[stations["가격"] != "-"]
stations.tail()
````

````python
#491 -> 489 개의 주유소가 출력이 되었다. (가격 미제공 주유소 2곳 제외)
stations["가격"] = stations["가격"].astype("float")
stations.info() #가격 타입이 문자열에서 float64로 변경되었다.
````

````python
stations
````

````python
# 인덱스 재정렬
stations.reset_index(inplace=True)
stations.tail()
````
